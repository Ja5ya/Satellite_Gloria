### Melt the Data:

```python
import pandas as pd
import re

# Function to parse the 'country_industry' column
def parse_labels(label):
    # Find the last set of parentheses for the country code
    match = re.search(r'\((\w{3})\)', label)
    
    if match:
        # Extract the country code
        country_code = match.group(1).strip()
        
        # Remove the country code part and trim the remaining string
        label_without_code = label[:match.start()].strip()
        industry_part = label[match.end():].strip()
        
        # Extract the country as everything before the last set of parentheses
        last_open_bracket_index = label_without_code.rfind('(')
        if last_open_bracket_index != -1:
            country = label_without_code[:last_open_bracket_index].strip()
        else:
            country = label_without_code.strip()
        
        # The remaining part after the country code is the industry
        industry = industry_part.strip()
        
        return pd.Series([country, country_code, industry], index=['Country', 'Country_Code', 'Industry'])
    else:
        # Handle cases where no country code is found
        return pd.Series([label, None, None], index=['Country', 'Country_Code', 'Industry'])

# Load your data
data = pd.read_csv('df_with_header.csv')

# Identify columns
id_vars = ['Year', 'Source_Filename']
value_vars = data.columns.difference(id_vars)

# Melt the data
long_data = pd.melt(data, id_vars=id_vars, value_vars=value_vars, 
                    var_name='country_industry', value_name='emissions')

# Apply the parse_labels function to split 'country_industry' into 'Country', 'Country_Code', and 'Industry'
parsed_labels = long_data['country_industry'].apply(parse_labels)
long_data = long_data.join(parsed_labels)

# Drop the 'country_industry' column
long_data = long_data.drop(columns=['country_industry'])

# Get distinct values for 'Country', 'Country_Code', and 'Industry'
distinct_countries = long_data['Country'].unique()
distinct_country_codes = long_data['Country_Code'].unique()
distinct_industries = long_data['Industry'].unique()

# Convert to DataFrame for better display
distinct_countries_df = pd.DataFrame(distinct_countries, columns=['Country'])
distinct_country_codes_df = pd.DataFrame(distinct_country_codes, columns=['Country_Code'])
distinct_industries_df = pd.DataFrame(distinct_industries, columns=['Industry'])

# Display the distinct values
print("Distinct Countries:")
print(distinct_countries_df)
print("\nDistinct Country Codes:")
print(distinct_country_codes_df)
print("\nDistinct Industries:")
print(distinct_industries_df)
```
